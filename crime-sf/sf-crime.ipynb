```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.options.display.max_columns = 50
```

```python
df = pd.read_csv('./Police_Department_Incident_Reports__2018_to_Present.csv')
```

```python
# filter for 2021
year = 2021
df = df[df['Incident Year'] == 2021]
```

```python
# preprocess
df['Incident Datetime'] = pd.to_datetime(df['Incident Datetime'])
```

```python
df.shape
```

```python
vehicle_theft_descriptions = \
df.loc[
    df['Incident Description'].str.lower().str.contains('vehicle') & \
    df['Incident Description'].str.lower().str.contains('theft'), 
'Incident Description'].unique()
```

```python
df = df[df['Incident Description'].isin(vehicle_theft_descriptions)]
```

```python
(df.isna().sum() / len(df) * 100).plot.bar()
```

```python
plt.figure(figsize=(15, 8))
order = df['Analysis Neighborhood'].value_counts().index
sns.countplot(data=df, x='Analysis Neighborhood', order=order);
plt.xticks(rotation=90);
```

Location seems to play an important part in determining the amount of break-ins. 

```python
df.head(2)
```

```python
df['Incident Hour'] = df['Incident Datetime'].dt.hour
```

```python
# 1. create hourly resolution datetime stamp
# 2. create hour timestamp column from that
# 3. df.groupby('hourly_timestamp').count() - use that to plot
df['Incident Datetime Hourly'] = df['Incident Datetime'].apply(lambda ts: ts.replace(minute=0, second=0))
```

```python
counts = df.groupby('Incident Datetime Hourly').count().iloc[:, 0]
counts = counts.reset_index()
counts = counts.rename(columns={counts.columns[1]: 'Count'})
counts['week'] = counts['Incident Datetime Hourly'].dt.week
plt.figure(figsize=(12, 6))
sns.boxplot(data=counts, x='week', y='Count', showfliers=False)
```

```python
counts = df.groupby('Incident Datetime Hourly').count().iloc[:, 0]
counts = counts.reset_index()
counts = counts.rename(columns={counts.columns[1]: 'Count'})
counts['Day'] = counts['Incident Datetime Hourly'].dt.dayofweek
plt.figure(figsize=(12, 6))
sns.boxplot(data=counts, x='Day', y='Count', showfliers=False)
```

```python
counts = df.groupby('Incident Datetime Hourly').count().iloc[:, 0]
counts = counts.reset_index()
counts = counts.rename(columns={counts.columns[1]: 'Count'})
counts['Hour'] = counts['Incident Datetime Hourly'].dt.hour
plt.figure(figsize=(12, 6))
sns.boxplot(data=counts, x='Hour', y='Count', showfliers=False)
```

Observations:
- Not much inter-group variability looking at week of year
- Not much inter-group variability looking at month of year
- Definite inter-group variability looking at month of year
